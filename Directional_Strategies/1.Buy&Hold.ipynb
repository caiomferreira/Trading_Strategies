```python
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots


from enum import Enum
from scipy.stats import norm
import pandas as pd
import numpy as np

import os
import sys

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
from _utils.core_functions import *

DATA_DIR = "../_databases"

DEFAULT_DATE_FORMAT = "%Y-%m-%d"

```

```python
df_win = pd.read_excel((DATA_DIR + '/WINFUT.xlsx'),decimal=',',index_col='date')
df_win.sort_index(inplace=True)

df_wdo = pd.read_excel((DATA_DIR + '/WDOFUT.xlsx'),decimal=',',index_col='date')
df_wdo.sort_index(inplace=True)
```

```python
adjusted_price = df_win["close"]
current_price = df_win["adjusted_close"]
multiplier = 0.2
fx_series = pd.Series(1, index=df_win.index)
position_contracts_held = pd.Series(100, index=df_win.index)
capital_required = multiplier * current_price

buyhold_win_returns = calculate_perc_returns(
    position_contracts_held, adjusted_price, fx_series, multiplier, capital_required
)
```

```python
adjusted_price = df_wdo["close"]
current_price = df_wdo["adjusted_close"]
multiplier = 10
fx_series = pd.Series(1, index=df_wdo.index)
position_contracts_held = pd.Series(50, index=df_wdo.index)
capital_required = multiplier * current_price

buyhold_wdo_returns = calculate_perc_returns(
    position_contracts_held, adjusted_price, fx_series, multiplier, capital_required
)

```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=buyhold_wdo_returns.index,
        y=buyhold_wdo_returns.cumsum(),
        name="WDOFUT - 50 contratos",
        line=dict(color="green"),
    )
)

fig.add_trace(
    go.Scatter(
        x=buyhold_win_returns.index,
        y=buyhold_win_returns.cumsum(),
        name="WINFUT - 100 contratos",
        line=dict(color="blue"),
    )
)

fig.update_layout(
    title="Buy & Hold Strategy - Retorno acumulado",
    xaxis_title="Data",
    yaxis_title="Retorno (%)",
    template="plotly_white",
    showlegend=True,
    height=600,
    width=1000,
)
fig.show()
```
